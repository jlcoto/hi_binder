```python
from itertools import cycle

import numpy as np

from bokeh.events import ButtonClick
from bokeh.io import output_notebook
from bokeh.layouts import widgetbox, row
from bokeh.models import ColumnDataSource, LabelSet
from bokeh.models.callbacks import CustomJS
from bokeh.models.widgets import Button
from bokeh.palettes import brewer
from bokeh.plotting import figure, show
```

```python
output_notebook()
```

```python
x = range(500)
y = range(500)
```

```python
cycled = cycle(brewer['Dark2'][3])
color_cycle = list(next(cycled) for _ in range(500))
```

```python
data_fake = {'x': x, 
            'y': y,
            'message': np.repeat('', len(y)),
            'color_list': color_cycle,
            'alpha': np.repeat(0.0, len(y))}
```

```python
source = ColumnDataSource(data_fake)
p = figure(sizing_mode = "scale_width", title="Say Hi to Binder")
p.circle('x', 'y', alpha='alpha', size=0, color='color_list', source=source)

callback = CustomJS(args=dict(source_1=source), code="""
            let ran_num_x = Math.floor(Math.random() * Math.floor(500));
            let ran_num_y = Math.floor(Math.random() * Math.floor(500));
            let data = source_1.data;
            let alpha = data['alpha']
            let message = data['message']
            data['x'][ran_num_x] = ran_num_x;
            data['y'][ran_num_x] = ran_num_y;
            alpha[ran_num_x] = 0.8
            message[ran_num_x] = 'Hi Binder :) !'
            source_1.change.emit();
""")


labels = LabelSet(x='x', y='y', text='message', text_color='color_list', level='glyph',
              x_offset=0, y_offset=0, source=source)

button = Button(label="Click here", button_type="success")

button.js_on_event(ButtonClick, callback)

p.add_layout(labels)
show(row(p, widgetbox(button), sizing_mode='scale_width'))
```
